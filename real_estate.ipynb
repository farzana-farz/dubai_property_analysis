```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("dubai_properties.csv")
data
```

```python
# count null values
data.isnull().sum()
```

```python
# counting unique values
data.nunique()
```

```python
#sort dataset by rent
data.sort_values(by='Rent').head(20)
```

```python
#count zeros 
count_rent = data['Rent'].value_counts()[0.00]
print("rent : ",count_rent)

count_beds = data['Beds'].value_counts()[0.00]
print("bed : ",count_beds)

count_rent_sqft = data['Rent_per_sqft'].value_counts()[0.00]
print("Rent_per_sqft : ",count_rent_sqft)
```

```python
#remove duplicates
data.drop_duplicates()
```

```python
#remove cols 'Age_of_listing_in_days','Purpose','Frequency'
data=data.drop(columns=['Age_of_listing_in_days','Purpose','Frequency'])
data
```

```python
#split address to 2 cols
data[['Building Name','Place']]=data['Address'].str.split(',',n=1,expand=True)
data
```

```python
#remove cols 'Address','Place'
data=data.drop(columns=['Address','Place','Latitude','Longitude'])
data
```

```python
# remove rows with value 0 in col Rent
data = data[data['Rent'] != 0]

data
```

```python
#checking null values
data.isna().sum()
```
